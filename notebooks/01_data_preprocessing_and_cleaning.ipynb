```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Load raw data from CSV via GitHub link
df = pd.read_csv('https://raw.githubusercontent.com/sngillard/student-outcome-classification/refs/heads/main/data/raw/student_data_raw.csv'
)

# Fix spelling error in column
df.rename(columns={"Nacionality" : "Nationality"}, inplace=True)

# View first 20 rows
df.head(20)
```

```python
# Check for missing values
df.info()

# None missing (4424 entries, indexes 0-4423
```

```python
# Describe the stats of the df (no missing values)
df.describe(include="all").transpose()
```

```python
le = LabelEncoder()
# Fit label encoder and return encoded labels
df["Target"] = le.fit_transform(df["Target"])
# Return count of unique values
df["Target"].value_counts()
```

```python
# Split into X (features) and y (target variable aka "Target" column which is drop out or 0, enrolled or 1, or graduate or 2)
X = df.drop(columns=['Target'], axis=1)
y = df['Target']

# Scale numeric features for logistic regression
std_scaler = StandardScaler()

# StandardScaler scales feature data and creates a NumPy array called X_features_scaled_array that contains the X/feature values
X_features_scaled_array = std_scaler.fit_transform(X)

# Turn scaled X_features_scaled_array into a Dataframe with original column names
X_scaled = pd.DataFrame(X_features_scaled_array, columns=X.columns)

# Make copy of the scaled features
df_preprocessed = X_scaled.copy()

# Add original target/y variable to scaled data frame
df_preprocessed['Target'] = y.values
'''
# Test
print(df_preprocessed.shape)
display(df_preprocessed.head())
display(df_preprocessed.describe().transpose())
'''

# Save processed dataset as new .csv file
df_preprocessed.to_csv('../student_data_processed_20251208.csv', index = False) # Drop column "Unnamed: 0" from table
```
